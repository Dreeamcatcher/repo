# Основные концепции Machine Learning. Домашняя работа

```python
### Тут нужно сделать импорты всех необходимых библиотек ###

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from itertools import combinations
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
matplotlib.rc("font", size=18)
```

```python
### Здесь требуется загрузить датасет phones.csv ###
phones = pd.read_csv("1.8_phones.csv", sep=",")
phones.head(5)
```

```python
### Опционально: используя matplotlib, можно сделать визуализацию загруженного датасета ###
f,ax=plt.subplots(1, 3, sharey=True, figsize=(20,10))

ax[0].scatter(phones['os'], phones['price'], s=100)
ax[1].scatter(phones['year'], phones['price'], s=100)
ax[2].scatter(phones['disk'], phones['price'], s=100)

ax[0].set_title("OS", fontsize=20)
ax[1].set_title("year", fontsize=20)
ax[2].set_title("disk", fontsize=20)

ax[2].set_xticks([16,32,64,128,256,512])

plt.tight_layout()
plt.show()
```

Задание 1: обучить модель регрессии

```python
### Тут нужно выгрузить в соответствующие переменные выбранные признаки ###
phones['os_id'] = phones['os'].apply(lambda x: 1 if x == 'Android' else 0)

X = phones[['year', 'os_id', 'disk']]
y = phones['price']
```

```python
### Здесь следует запустить процесс обучения ###
### и продемонстрировать работоспобность обученной модели ###
reg = LinearRegression().fit(X,y)
```

```python
print('2016 android 32gb - ' , reg.predict([[2016, 1, 32]])[0])
print('2016 iOS 32gb - ' , reg.predict([[2016, 0, 32]])[0])
print('2021 android 512gb - ' , reg.predict([[2021, 1, 512]])[0])
print('2021 iOS 512gb - ' , reg.predict([[2021, 0, 512]])[0])
```

Задание 2: обучить модель классификации

```python
### Тут нужно выгрузить в соответствующие переменные выбранные признаки ###
X = phones[['price', 'year']]
y = phones['os']
```

```python
### Здесь следует запустить процесс обучения ###
### и продемонстрировать работоспобность обученной модели ###
cl = DecisionTreeClassifier().fit(X, y)
print(export_text(cl))
```

```python
for i in [np.random.randint(0, 100) for _ in range(5)] :
    print(X[i:i+1],' predict - ', cl.predict(X[i:i+1])[0], ' | real - ', y[i] )
```

Задание 3: обучить модель кластеризации

```python
X = phones[['year', 'disk', 'price']]
```

Для того чтобы оптимально подобрать кол-во кластеров для нашей выборки, воспользуемся методами локтя и коэффициента силуэта.

```python
km_scores = []
km_silhouette = []

# обучаем алгоритм с различным количеством кластеров, и оцениваем
for i in range(2,12):
    km = KMeans(n_clusters=i, random_state=0).fit(X)
    preds = km.predict(X)
    # метод локтя
    km_scores.append(-km.score(X))
    # коэфициент силуэта
    silhouette = silhouette_score(X,preds)
    km_silhouette.append(silhouette)

plt.figure(figsize=(7,4))
plt.title("Определение кол-ва кластеров методом локтя", fontsize=16)
plt.plot([i for i in range(2,12)], km_scores, 'b*-')
plt.xlabel("Количество кластеров", fontsize=14)
plt.ylabel("K-means score", fontsize=15)
plt.xticks([i for i in range(2,12)], fontsize=14)
plt.grid(True)
plt.show()
```

По методу локтя оптимальное кол-во кластеров - 3, 5 или 6

```python
plt.figure(figsize=(7,4))
plt.title("Определение кол-ва кластеров  методом коэффициента силуэта", fontsize=16)
plt.plot([i for i in range(2,12)], km_silhouette,  'b*-')
plt.xlabel("Количество кластеров", fontsize=14)
plt.ylabel("Silhouette score", fontsize=15)
plt.xticks([i for i in range(2,12)], fontsize=14)
plt.grid(True)
plt.show()
```

По методу коэффициента силуэта оптимальное кол-во кластеров - 6.

**По совокупности двух методов, будем считать количество кластеров равным 6**

```python
n_clusters = 6

clusters = KMeans(n_clusters=n_clusters).fit(X)


phones['cluster'] = clusters.labels_
phones.head(10)
```

Посмотрим полученные кластеры в разрезе по каждой паре признаков:

```python
# получаем возможные комбинации фичей
combs = list(combinations(range(0,3), 2))

f,ax = plt.subplots(1, len(combs), figsize=(20,5))

for i in range(0, len(combs)):
    dim1 = combs[i][0]
    dim2 = combs[i][1]
    ax[i].scatter(X.iloc[:,dim1], X.iloc[:,dim2], s=100, alpha=0.3)
    ax[i].set_xlabel(f"{X.columns[dim1]}")
    ax[i].set_ylabel(f"{X.columns[dim2]}")

    # кластеры
    for cl in clusters.cluster_centers_:
        ax[i].scatter(cl[dim1], cl[dim2], s=250, marker='x', linewidths=3, color='black')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
ax = plt.axes(projection='3d')

# для наглядности отметим виды ОС разными цветами
colors = {'Android':'green', 'iOS':'red'}
for item in colors.keys():
    df = phones[phones.os == item]
    ax.scatter(df.year, df.disk, df.price, s=100, color=colors[item])

# обозначаем центры кластеров
for cl in clusters.cluster_centers_:
    ax.scatter(cl[0], cl[1], cl[2], s=250 , marker='x', color='black')

ax.set_xlabel("year")
ax.set_ylabel("disk")
ax.set_zlabel("price")
ax.set_yticks([16,32,64,128,256,512])
ax.tick_params(labelsize=10)
# поворачиваем график
ax.view_init(15, 170)
plt.show()
```
